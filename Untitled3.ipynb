```python
# Paso 1: Importar librerías
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Configuraciones para que se vean bien los gráficos
sns.set(style="whitegrid")
plt.rcParams["figure.figsize"] = (12,6)


# Asumiendo que el archivo se llama igual al que subiste
url="https://raw.githubusercontent.com/saith20/electiva/refs/heads/main/Inscripciones_Actividad_f_sica_y_Deportes_20250416.csv "
df = pd.read_csv(url)
df.head()

# Paso 3: Exploración básica
print("Primeras filas del dataset:")
display(df.head())

print("\nInformación general del dataset:")
df.info()

print("\nResumen estadístico:")
display(df.describe(include='all'))

print("\nValores nulos por columna:")
print(df.isnull().sum())

# Paso 4: Análisis básico
# Por ejemplo: distribución de inscripciones por actividad (si hay una columna de actividad)
if 'Actividad' in df.columns:
    plt.figure(figsize=(10,6))
    sns.countplot(data=df, x='Actividad', order=df['Actividad'].value_counts().index)
    plt.title("Distribución de Inscripciones por Actividad")
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()

# O distribución por género, si hay columna 'Sexo' o 'Género'
for col in ['Sexo', 'Género']:
    if col in df.columns:
        plt.figure(figsize=(6,4))
        df[col].value_counts().plot(kind='bar', color='skyblue')
        plt.title(f'Distribución por {col}')
        plt.xlabel(col)
        plt.ylabel('Cantidad')
        plt.show()



# Puedes seguir agregando análisis más específicos según el contenido del dataset

# Gráfica: Top 10 deportes o servicios más inscritos
if 'Deporte o Servicio' in df.columns:
    top_deportes = df['Deporte o Servicio'].value_counts().nlargest(10)
    plt.figure(figsize=(10,6))
    sns.barplot(x=top_deportes.values, y=top_deportes.index, palette="muted")
    plt.title("Top 10 Deportes o Servicios con Mayor Número de Inscripciones")
    plt.xlabel("Cantidad de Inscripciones")
    plt.ylabel("Deporte o Servicio")
    plt.tight_layout()
    plt.show()

# Gráfica: Distribución de inscripciones por edad (en bins)
if 'Edad' in df.columns:
    plt.figure(figsize=(10,6))
    sns.histplot(df['Edad'], bins=15, kde=True, color="coral")
    plt.title("Distribución de Inscripciones por Edad")
    plt.xlabel("Edad")
    plt.ylabel("Frecuencia")
    plt.tight_layout()
    plt.show()




```
